### Sentiment Analysis for Review Columns

For sentiment analysis, we will be using the pretrained [bert-base-multilingual-uncased-sentiment](https://huggingface.co/nlptown/bert-base-multilingual-uncased-sentiment) sentiment analysis model.


```python
from transformers import pipeline
```

```python
# combine review title and review content
df['review_body'] = df['review_title'] + ': ' + df['review_content']
df.head()
```

```python
# Let's see what it looks like

df['review_body'][0]
```

```python
# Implement truncate function to truncate any text entry larger than sentiment model input requirement

def truncate_text(text, max_length=512):
    return text[:max_length]
```

```python
# apply sentiment analysis to review content
pipe = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
df['review_sentiment'] = df['review_body'].apply(lambda x: pipe(truncate_text(x))[0]['label'])

# transform sentiment to numeric using Ordinal Encoder
oe = OrdinalEncoder(categories = [['5 stars', '4 stars', '3 stars', '2 stars', '1 star']])
df['review_sentiment_num'] = oe.fit_transform(df[['review_sentiment']])

df.head()
```

## Training the Model

We will train a Linear Regression Model, Random Forest Regressor Model, and a XGBoost Model and then compare results. For model evaluation we will be using mean squared error and r2 scores.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# category	discounted_price	actual_price	rating	rating_count	selling_price	review_sentiment	review_sentiment_num

Y = df['rating'].values
X = df[['category', 'discounted_price', 'actual_price', 'rating_count', 'selling_price', 'review_sentiment_num']].values
```

```python
# train test split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

Training With Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_pred = rfr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

Training with XGBoost

```python
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

Conslusion

Based on the mean squared error, the random forest come out front with a mse of 0.0795. Meanwhile, based on the R-Squared error, the XGBoost model comes out on top with a r2 score of 0.150.
